```python
myimagebilinear = misc.imresize(myimage.astype('uint8'),[1000, 1000], interp='bilinear')


myimagenearest = misc.imresize(myimage.astype('uint8'),[1000, 1000], interp='nearest')

myimagebicubic = misc.imresize(myimage.astype('uint8'),[1000, 1000], interp='bicubic')



fig, axs = plt.subplots(nrows=3,ncols= 1, figsize=(20, 20))
ax0, ax1,ax2=axs.ravel()

ax0.imshow(myimagebilinear, cmap=cm.gray)
ax0.set_title('BILINEAR')

ax1.imshow(myimagenearest, cmap=cm.gray)
ax1.set_title('NEAREST')

ax2.imshow(myimagebicubic, cmap=cm.gray)
ax2.set_title('BICUBIC')
```

** Using your functions, blend `orange` and `apple` images together.**


```python
file1 = './images/apple.jpeg'
apple=io.imread(file1, as_grey=True)
file2 = './images/orange.jpeg'
orange=io.imread(file2, as_grey=True)

```

```python

def blendalpha(apple, orange,alpha):
    
    return (alpha * apple + (1 - alpha) * orange)
    
```

```python
simple=blendalpha(apple,orange,0.5)
io.imshow(simple)
```
